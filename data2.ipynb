```python
import requests
import pandas as pd
```

```python
twitter_data = []
```

```python
payload1 = {
    'api_key':'',
    'query':'Israel War',
    'num':'100'
}

payload2 = {
    'api_key':'',
    'query':'Israel',
    'num':'100'
}

payload3 = {
    'api_key':'',
    'query':'Israel Hamas War',
    'num':'100'
}

payload4 = {
    'api_key':'',
    'query':'Gaza',
    'num':'100'
}
```

```python
response1 = requests.get(
    'https://api.scraperapi.com/structured/twitter/search', params=payload1
)
data1 = response1.json()
data1
```

```python
response2 = requests.get(
    'https://api.scraperapi.com/structured/twitter/search', params=payload2
)
data2 = response2.json()
data2
```

```python
response3 = requests.get(
    'https://api.scraperapi.com/structured/twitter/search', params=payload3
)
data3 = response3.json()
data3
```

```python
response4 = requests.get(
    'https://api.scraperapi.com/structured/twitter/search', params=payload4
)
data4 = response4.json()
data4
```

```python
data1.keys()
```

```python
all_tweets = data1['organic_results']+data2['organic_results']+data3['organic_results']+data4['organic_results']
for tweet in all_tweets:
  twitter_data.append(tweet)
```

```python
df = pd.DataFrame(twitter_data)
df
```

```python
from matplotlib import pyplot as plt
df['position'].plot(kind='hist', bins=20, title='position')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
df['position'].plot(kind='line', figsize=(8, 4), title='position')
plt.gca().spines[['top', 'right']].set_visible(False)
```
