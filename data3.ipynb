```python
import requests
import pandas as pd
```

```python
all_tweets = []
twitter_data = []
```

```python
keywords = ['Israel', 'Hamas', 'Palestine', 'Israel War', 'Hamas War', 'Palestine War', 'Israel vs Palestine', 'Palestine vs Israel', 'Gaza', 'Gaza Attacks']
```

```python
for keyword in keywords:
  payload = {
      'api_key':'',
      'query':keyword,
      'num':'200'
  }

  response = requests.get(
      'https://api.scraperapi.com/structured/twitter/search', params=payload
  )

  data = response.json()
  all_tweets = all_tweets + data['organic_results']
```

```python
len(all_tweets)
```

```python
for tweet in all_tweets:
  twitter_data.append(tweet)
```

```python
df = pd.DataFrame(twitter_data)
df
```
